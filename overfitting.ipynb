```python
import matplotlib.pyplot as plt

# 仮の学習履歴データ (実際の学習では model.fit() のhistoryオブジェクトから取得)
epochs = range(1, 21)
train_loss = [0.8, 0.6, 0.45, 0.35, 0.28, 0.22, 0.18, 0.15, 0.13, 0.11, 0.10, 0.09, 0.08, 0.075, 0.07, 0.068, 0.067, 0.066, 0.065, 0.064]
val_loss =   [0.9, 0.7, 0.55, 0.48, 0.42, 0.38, 0.35, 0.33, 0.32, 0.31, 0.315, 0.32, 0.33, 0.34, 0.35, 0.36, 0.37, 0.38, 0.39, 0.40]

plt.figure(figsize=(8, 5))
plt.plot(epochs, train_loss, label='Training Loss', color='blue')
plt.plot(epochs, val_loss, label='Validation Loss', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over Epochs (Identifying Overfitting)')
plt.legend()
plt.grid(True)
plt.show()
```
